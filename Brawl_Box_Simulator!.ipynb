```python
# @title
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

rarities = ['(Rare)', '(Super rare!)', '(Epic!!)', '(Mythic!!!!)', '(Legendary!!!!!!!!!!!!)', 'power points and coins', 'gadget', 'starpower']
common_brawlers = ['Shelly!!! (Starting brawler)']
rare_brawlers = ['Nita', 'Colt', 'Bull', 'Brock', 'El Primo', 'Barley', 'Poco', 'Rosa']
super_rare_brawlers = ['Jessie', 'Dynamike', 'Tick', '8-Bit', 'Rico', 'Darryl', 'Penny', 'Carl', 'Jacky', 'Gus']
epic_brawlers = ['Bo', 'Emz', 'Stu', 'Piper', 'Pam', 'Frank', 'Bibi', 'Bea', 'Nani', 'Edgar', 'Griff', 'Grom', 'Bonnie', 'Gale', 'Colette', 'Belle', 'Ash', 'Lola', 'Sam', 'Mandy', 'Maisie', 'Hank', 'Pearl', 'Larry & Lawrie', 'Angelo']
mythic_brawlers = ['Motis', 'Tara', 'Gene', 'Max', 'Mr.P', 'Sprout', 'Byron', 'Squeak', 'Lou', 'Ruffs', 'Buzz', 'Fang', 'Eve', 'Janet', 'Otis', 'Buster', 'Gray', 'R-T', 'Willow', 'Doug', 'Chuck', 'Charlie', 'Mico', 'Melodie', 'Lily']
legendary_brawlers = ['Spike', 'Crow', 'Leon', 'Sandy', 'Amber', 'Meg', 'Surge', 'Chester', 'Cordelius', 'Kit']
attempt = 0
repeats = 0

rare_brawlers = [elem + ' ' + rarities[0] for elem in rare_brawlers]
super_rare_brawlers = [elem + ' ' + rarities[1] for elem in super_rare_brawlers]
epic_brawlers = [elem + ' ' + rarities[2] for elem in epic_brawlers]
mythic_brawlers = [elem + ' ' + rarities[3] for elem in mythic_brawlers]
legendary_brawlers = [elem + ' ' + rarities[4] for elem in legendary_brawlers]

all_brawlers = rare_brawlers + super_rare_brawlers + epic_brawlers + mythic_brawlers + legendary_brawlers

##rare, super rare, epic, mythic, legendary, power points and coins, gadget, starpower
rarity_weights = [.0277, .0125, .0056, .0026, .0011, .9205, .02, .01]

open_box = widgets.Button(description = 'Test my odds!')
open_box.layout.width = 'auto'
open_box.layout.height = 'auto'
open_box.style.button_color = 'green'
open_box.style.font_color = 'black'

reset_button = widgets.Button(description = 'Reset brawlers')
reset_button.layout.width = 'auto'
reset_button.layout.height = 'auto'
reset_button.style.button_color = 'red'

repeats = 0
brawlers_picked = []
output = widgets.Output()
brawler = []
coins_one_try = 0
coins = 0
coins_per_round = 0
p_points_one_try = 0
p_points = 0
p_points_per_round = 0
brawlers_picked_init = []
gadget_brawlers = []
avail_gadget_brawlers = []
star_power_brawlers = []
avail_star_power_brawlers = []

def click(open_box):
  global brawlers_picked, rarities, rare_brawlers, super_rare_brawlers, epic_brawlers, mythic_brawlers, legendary_brawlers, attempt, repeats, brawler, tries, coins, coins_one_try, p_points, p_points_one_try, brawlers_picked_init, coins_per_round, p_points_per_round, gadget_brawlers, star_power_brawlers, avail_gadget_brawlers, avail_star_power_brawlers

##Choosing a box
  boxes = ['brawl box', 'big box', 'MEGA box']
  box_weights = [.3,.5,.2]
  box = random.choices(boxes, weights=box_weights, k=1)[0]
  if box == 'brawl box':
    tries = 3
    open_box.style.button_color = 'blue'
  elif box == 'big box':
    tries = 5
    open_box.style.button_color = 'purple'
  elif box == 'MEGA box':
    tries = 10
    open_box.style.button_color = 'yellow'
  else:
    tries = 0
  repeats +=tries

##If there aren't any brawlers left
  if len(brawlers_picked) == len(common_brawlers + rare_brawlers + super_rare_brawlers + epic_brawlers + mythic_brawlers + legendary_brawlers):
    with output:
      clear_output()
      print("Not enough brawlers left! You've got so many already :D")
    return print(f"You have:", brawlers_picked)

##Chosing a random brawler based on the amount of tries
  else:
    attempt += 1
    if attempt == 1:
      brawler = random.choices(common_brawlers,k=1)
      brawlers_picked_init = list(set(brawlers_picked_init + brawler))
      gadget_brawlers = list(set(gadget_brawlers + brawler + brawler))
      star_power_brawlers = list(set(star_power_brawlers + brawler + brawler))
    for i in range(tries):
      rarity = random.choices(rarities, weights=rarity_weights, k=1)[0]
      if rarity == rarities[0]: ##Rare
        if len(rare_brawlers) == 0:
          repeats += 1
          coins_p_points(1)
        else:
          brawler = random.choices(rare_brawlers,k=1)
          brawlers_picked_init = list(set(brawlers_picked_init + brawler))
          gadget_brawlers = list(set(gadget_brawlers + brawler + brawler))
          star_power_brawlers = list(set(star_power_brawlers + brawler + brawler))
          common_brawler = None
          for item in rare_brawlers:
            if item in brawler:
              common_brawler = item
              break
          if common_brawler:
            rare_brawlers.remove(common_brawler)
      elif rarity == rarities[1]: ##Super rare
        if len(super_rare_brawlers) == 0:
          repeats += 1
          coins_p_points(1)
        else:
          brawler = random.choices(super_rare_brawlers, k=1)
          brawlers_picked_init = list(set(brawlers_picked_init + brawler))
          gadget_brawlers = list(set(gadget_brawlers + brawler + brawler))
          star_power_brawlers = list(set(star_power_brawlers + brawler + brawler))
          common_brawler = None
          for item in super_rare_brawlers:
            if item in brawler:
              common_brawler = item
              break
          if common_brawler:
            super_rare_brawlers.remove(common_brawler)
      elif rarity == rarities[2]: ##Epic
        if len(epic_brawlers) == 0:
          repeats += 1
          coins_p_points(1)
        else:
          brawler = random.choices(epic_brawlers, k=1)
          brawlers_picked_init = list(set(brawlers_picked_init + brawler))
          gadget_brawlers = list(set(gadget_brawlers + brawler + brawler))
          star_power_brawlers = list(set(star_power_brawlers + brawler + brawler))
          common_brawler = None
          for item in epic_brawlers:
            if item in brawler:
              common_brawler = item
              break
          if common_brawler:
            epic_brawlers.remove(common_brawler)
      elif rarity == rarities[3]: ##Mythic
        if len(mythic_brawlers) == 0:
          repeats += 1
          coins_p_points(1)
        else:
          brawler = random.choices(mythic_brawlers, k=1)
          brawlers_picked_init = list(set(brawlers_picked_init + brawler))
          gadget_brawlers = list(set(gadget_brawlers + brawler + brawler))
          star_power_brawlers = list(set(star_power_brawlers + brawler + brawler))
          common_brawler = None
          for item in mythic_brawlers:
            if item in brawler:
              common_brawler = item
              break
          if common_brawler:
            mythic_brawlers.remove(common_brawler)
      elif rarity == rarities[4]: ##Legendary
        if len(legendary_brawlers) == 0:
          repeats += 1
          coins_p_points(1)
        else:
          brawler = random.choices(legendary_brawlers, k=1)
          brawlers_picked_init = list(set(brawlers_picked_init + brawler))
          gadget_brawlers = list(set(gadget_brawlers + brawler + brawler))
          star_power_brawlers = list(set(star_power_brawlers + brawler + brawler))
          common_brawler = None
          for item in legendary_brawlers:
            if item in brawler:
              common_brawler = item
              break
          if common_brawler:
            legendary_brawlers.remove(common_brawler)
      elif rarity == rarities[5]: ##Power points and coins
        coins_per_round = 0
        coins_one_try = random.randint(20,50)
        coins_per_round +=coins_one_try
        coins += coins_one_try
        p_points_per_round = 0
        p_points_one_try = random.randint(10,40)
        p_points_per_round += p_points_one_try
        p_points += p_points_one_try
      elif rarity == rarities[6]: ##Gadget
        if gadget_brawlers:
          gadget_brawler = random.choices(gadget_brawlers,k=1)
          print(f"You got a gadget for {gadget_brawler}!")
          common_gadget_brawler = None
          for item in gadget_brawlers:
            if item in gadget_brawler:
              common_gadget_brawler = item
              break
          if common_gadget_brawler:
            gadget_brawlers.remove(common_gadget_brawler)
        else:
          coins_p_points(1)
      elif rarity == rarities[7]: ##Star power
        if star_power_brawlers:
          star_power_brawler = random.choices(star_power_brawlers,k=1)
          print(f"You got a star power for {star_power_brawler}!")
          common_star_power_brawler = None
          for item in star_power_brawlers:
            if item in star_power_brawler:
              common_star_power_brawler = item
              break
          if common_star_power_brawler:
            star_power_brawlers.remove(common_star_power_brawler)
        else:
          coins_p_points(1)

    open_box.description = 'Try again?'
    if brawlers_picked != brawlers_picked_init:
      brawlers_picked = list(set(brawlers_picked + brawlers_picked_init))
      with output:
        clear_output()
        print(f'\nYaay!!\nAttempt: {attempt}\nTotal draws: {repeats}\nRepeated draws: {repeats-len(brawlers_picked)}\nCoins: {coins}\nPower points: {p_points}\nBrawlers you have: {len(brawlers_picked)}\nBrawlers left to get: {len(all_brawlers)-len(brawlers_picked)}\n\nYou opened a {box} and now have:',brawlers_picked)
        print("")
    else:
      brawlers_picked = list(set(brawlers_picked + brawlers_picked_init))
      with output:
        clear_output()
        print(f"\nYaay!!\nAttempt: {attempt}\nTotal draws: {repeats}\nRepeated draws: {repeats-len(brawlers_picked)}\nCoins: {coins}\nPower points: {p_points}\nBrawlers you have: {len(brawlers_picked)}\nBrawlers left to get: {len(all_brawlers)-len(brawlers_picked)}\n\nSorry, you didn't get any new brawlers this try, but you got {coins_per_round} coins and {p_points_per_round} power points!\n\nYou opened a {box}, but still have:",brawlers_picked)
        print("")

def reset(idontevenknowwhatthisfieldisfor):
  global brawlers_picked, attempt, repeats, star_power_brawlers, gadget_brawlers
  if brawlers_picked:
    del brawlers_picked
    del attempt
    del repeats
    del star_power_brawlers
    del gadget_brawlers
    open_box.description = 'Test my odds again!? Open box!'
    open_box.style.button_color = 'green'
    with output:
      clear_output()
      print('Brawlers reset. If you would like to try your luck again, press "Open box"')
      brawlers_picked = []
      brawler = []
      star_power_brawlers = []
      gadget_brawlers = []
      attempt = 0
      repeats = 0
  else:
    with output:
      clear_output()
      print("Nothing to reset!")

def coins_p_points(b):
  global coins,p_points
  coins_per_round = 0
  coins_one_try = random.randint(20,50)
  coins_per_round +=coins_one_try
  coins += coins_one_try
  p_points_per_round = 0
  p_points_one_try = random.randint(10,40)
  p_points_per_round += p_points_one_try
  p_points += p_points_one_try

open_box.on_click(click)
reset_button.on_click(reset)
display(open_box, output)
```
